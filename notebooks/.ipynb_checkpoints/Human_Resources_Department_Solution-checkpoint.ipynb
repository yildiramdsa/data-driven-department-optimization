# VISUALIZE DATASET

```python
# As 'Attrition,' 'OverTime,' and 'Over18' are Yes/No columns,
# let's replace them with integers before performing any visualizations.
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x == 'Y' else 0)
```

```python
employee_df.head(3)
```

```python
# Let's check to see if we have any missing data. Fortunately, we don't have any!
sns.heatmap(employee_df.isnull(), yticklabels = False, cbar = False, cmap='Blues');
```

```python
employee_df.hist(bins=30, figsize=(20,20), color='blue');
# Several features like 'MonthlyIncome' and 'TotalWorkingYears' have a tail-heavy distribution.
```

```python
# Let's drop 'EmployeeCount,' 'StandardHours,' and 'Over18' since they do not vary among employees.
# Also, let's drop 'EmployeeNumber' because it doesn't provide meaningful information or patterns to help predict a target variable.
# Including such identifiers in a model can lead to overfitting, where the model memorizes specific employees rather than learning generalizable patterns.
employee_df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
employee_df.head(3)
```

```python
# Let's find out how many employees have left the company.
employees_left = employee_df[employee_df['Attrition'] == 1]
employees_stayed = employee_df[employee_df['Attrition'] == 0]
```

```python
# Calculate the total number of employees
total_employees = len(employee_df)

# Calculate the number and percentage of employees who left the company
num_left = len(employees_left)
pct_left = (num_left / total_employees) * 100

# Calculate the number and percentage of employees who stayed with the company
num_stayed = len(employees_stayed)
pct_stayed = (num_stayed / total_employees) * 100

print(f"Total number of employees: {total_employees}")
print(f"Number of employees who left the company: {num_left} ({pct_left:.2f}%)")
print(f"Number of employees who stayed with the company: {num_stayed} ({pct_stayed:.2f}%)")

# Check for dataset imbalance
if pct_left != pct_stayed:
    print("Warning: The dataset is imbalanced.")
else:
    print("The dataset is balanced.")
```

```python
employees_left.describe()
```

```python
employees_stayed.describe()
# Let's compare the average and standard deviation of the employees who stayed and those who left:
# 'Age': The average age of the employees who stayed is higher than that of those who left.
# 'DailyRate': The daily rate of the employees who stayed is higher.
# 'DistanceFromHome': Employees who stayed live closer to work.
# 'EnvironmentSatisfaction' & 'JobSatisfaction': Employees who stayed generally have higher environment and job satisfaction.
# 'StockOptionLevel': Employees who stayed tend to have a higher stock option level.
```

# TRAIN AND EVALUATE A DEEP LEARNING MODEL

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500, activation='relu', input_shape=(50, )))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=500, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
# oversampler = SMOTE(random_state=0)
# smote_train, smote_target = oversampler.fit_sample(X_train, y_train)
# epochs_hist = model.fit(smote_train, smote_target, epochs = 100, batch_size = 50)
epochs_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 50)
```

```python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```
